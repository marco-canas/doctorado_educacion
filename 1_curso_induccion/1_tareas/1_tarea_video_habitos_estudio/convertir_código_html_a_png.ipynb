```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import os

def html_to_png(html_file_path, output_png_path, width=1100):
    """
    Convierte un archivo HTML a PNG usando Selenium
    html_file_path: ruta al archivo .html
    output_png_path: ruta de salida .png
    width: ancho de la ventana (px)
    """
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Sin ventana visible
    chrome_options.add_argument(f"--window-size={width},1080")
    
    driver = webdriver.Chrome(options=chrome_options)
    
    # Cargar archivo local
    driver.get(f"file://{os.path.abspath(html_file_path)}")
    
    # Esperar a que cargue completamente
    time.sleep(1)
    
    # Obtener altura real del contenido
    height = driver.execute_script("return document.body.scrollHeight")
    driver.set_window_size(width, height)
    time.sleep(0.5)
    
    # Capturar pantalla
    driver.save_screenshot(output_png_path)
    driver.quit()
    
    print(f"✅ Imagen guardada en: {output_png_path}")
    return output_png_path
```

```python

# --- USO ---
# Guarda tu HTML como 'infografia_final.html' y ejecuta:
html_to_png('infografia_final.html', 'infografia_final.png')
```
