# Introducción  

En esta asignatura exploremos las estrategias y herramientas digitales fundamentales para la innovación y mejora de los procesos de enseñanza-aprendizaje en diversos contextos educativos. 

Abordaremos conceptos clave de la educación digital, las tecnologías educativas emergentes y los modelos pedagógicos que sustentan la implementación de entornos de aprendizaje en línea (e-Learning), móvil (m-Learning) y mixto (b-Learning).



Aprenderemos a evaluar diferentes escenarios educativos para diseñar estrategias digitales pertinentes, incorporando herramientas y plataformas adecuadas. 

Además, se profundizaremos en los principios del diseño instruccional para la creación de entornos de aprendizaje digital efectivos y atractivos, utilizando herramientas de autor (LMS y CMS) y explorando técnicas como la gamificación y realidad virtual/aumentada.

 



Finalmente, analizaremos los desafíos y tendencias actuales de la educación digital, incluyendo la inteligencia artificial y el aprendizaje personalizado, así como las competencias y brechas digitales.
